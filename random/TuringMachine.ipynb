```python
class TM():
    
    #You can give the Turing Machine a name
    def __init__ (self,name="The Amazing Turing Machine"):
        self.name=name
  
    #Initialize a Turing Machine with a initial value and a length
    def create(self, value ,length=50):
        self.length=length
        self.value=value
        y=[' ']*self.length
        x=[' ']
        for z in self.value:
            x.append(z)
        x+=y
        return x
   
    

    #Print the values of the Turing Machine up to The initial value + values, how many extra blanks
    def printValues(self,tm,underlined,values=2):
        self.tm=tm+[' ']*5
        self.underlined=underlined
        self.values=values
        index=0
        x=[]
        for idx, val in enumerate(self.tm):
            if val != ' ':
                index=idx
            elif val==' ':
                pass
        for ix, va in enumerate(self.tm):
            if ix<underlined:
                x.append(va)
            elif ix==underlined:
                x.append("\033[4m{}\033[0m".format(va))
            elif ix>underlined and ix<=index+values:
                x.append(va)       
        print ('[%s]' % ', '.join(map(str, x)))
   
    #Get the symbols for the Turing Machine
    def symbols(self,vals):
        self.vals=vals 
        newsymbol=''
        currentSymbols=[' ']
        for y in str(self.vals):
            if y not in currentSymbols:
                currentSymbols.append(y)
        moreSymbols=True
        print("The current symbols are", currentSymbols)
        while(moreSymbols):
            requestSymbol=input("Would you like to add another symbol? y/n ")
            while(requestSymbol.lower() not in ['y','n']):
                requestSymbol=input("Would you like to add another symbol? y/n ")
            if requestSymbol.lower() =='n':
                moreSymbols=False
            elif requestSymbol.lower()=='y':
                newSymbol=input("Add another symbol(1st char) ")
                if newSymbol[0] not in currentSymbols:
                    currentSymbols.append(newSymbol[0])
                else:
                    pass
        print("Your symbols are: ", currentSymbols)
        return currentSymbols
    
    #Get the states for the Turing Machine
    def states(self):
        newstate=''
        currentStates=['Start']
        moreStates=True
        print("The current states are", currentStates)
        while(moreStates):
            requestState=input("Would you like to add another state? y/n ")
            while(requestState.lower() not in ['y','n']):
                requestState=input("Would you like to add another state? y/n ")
            if requestState.lower() =='n':
                moreStates=False
            elif requestState.lower()=='y':
                newState=input("Add another state: ")
                if newState not in currentStates:
                    currentStates.append(newState)
                else:
                    pass       
        print("Your states are: ", currentStates)
        return currentStates

    #Define the rules of this Turing Machine
    def rules(self,symbols,states):
        self.symbols=symbols
        self.states=states
        currentRules=[]
        print("The current rules are:", currentRules)
        Rules={}
        ruleNum=1
        for sta in self.states:
            for sym in self.symbols:
                newSymbol=input("State: {}, Symbol '{}': new symbol? ".format(sta,sym))
                while(newSymbol not in self.symbols):
                    print("Unavailable symbol")
                    newSymbol=input("State: {}, Symbol '{}': new symbol? ".format(sta,sym))                    
                newState=input("State: {}, Symbol '{}': new state? ".format(sta,sym))
                while(newState not in self.states):
                    print("Unavailable state")
                    newState=input("State: {}, Symbol '{}': new state? ".format(sta,sym))
                move=input("State: {}, Symbol: '{}': Left or right? (L/R) ".format(sta,sym))
                while(move[0] not in ['L','R','0']):
                    move=input("State: {}, Symbol: '{}': Left or right? (L/R) ".format(sta,sym))
                currentRules.append(newState)
                currentRules.append(newSymbol)
                currentRules.append(move[0])
                Rules["{} {}".format(str(sta),str(sym))]=currentRules
                currentRules=[]
                ruleNum+=1
        return Rules
   
    #Get the new state depending on rules
    def newState(self,state,symbol,rules):
        self.state=state
        self.symbol=symbol
        self.rules=rules
        rule= rules.get("{} {}".format(str(self.state),str(self.symbol)))
        return rule[0]
    
    #Get the new symbol depending on rules
    def newSymbol(self,state,symbol,rules):
        self.state=state
        self.symbol=symbol
        self.rules=rules
        rule=rules.get("{} {}".format(str(self.state),str(self.symbol)))
        return rule[1]
    
    #Move depending on rules
    def move(self,state,symbol,rules):
        self.state=state
        self.symbol=symbol
        self.rules=rules
        rule=rules.get("{} {}".format(str(self.state),str(self.symbol)))
        if rule[2] =='L':
            return -1
        elif rule[2] =='R':
            return 1
        elif rule[2] =='0':
            return 0
        
    #Get the Accepted States of the Turing Machine
    def acceptStates(self,rules):
        self.rules=rules
        ln=list(self.rules.keys())
        count=0
        states=[]
        for x in self.rules:
            count+=1
            print("Rule {} is: {}".format(count,x))
        moreStates=True
        while(moreStates):
            requestState=input("Would you like to add an accept state? y/n ")
            while(requestState.lower() not in ['y','n']):
                requestState=input("Would you like to an accept state? y/n ")
            if requestState.lower() =='n':
                moreStates=False
            elif requestState.lower()=='y':
                accept=int(input("What rule is your accept state? [1-{}]".format(count)))
                while (accept not in range(1,count)):
                    accept=int(input("What rule is your accept state? [1-{}]".format(count)))
                if accept not in states:
                    states.append(ln[accept-1])
                else:
                    pass       
        return states
    
    #Get the Rejected States of the Turing Machine
    def rejectStates(self,rules):
        self.rules=rules
        ln=list(self.rules.keys())
        count=0
        states=[]
        for x in self.rules:
            count+=1
            print("Rule {} is: {}".format(count,x))
        moreStates=True
        while(moreStates):
            requestState=input("Would you like to add a reject state? y/n ")
            while(requestState.lower() not in ['y','n']):
                requestState=input("Would you like to an reject state? y/n ")
            if requestState.lower() =='n':
                moreStates=False
            elif requestState.lower()=='y':
                reject=int(input("What rule is your reject state? [1-{}]".format(count)))
                while (reject not in range(1,count)):
                    reject=int(input("What rule is your reject state? [1-{}]".format(count)))
                if reject not in states:
                    states.append(ln[reject-1])
                else:
                    pass       
        return states
    
    def haltState(self,rules):
        self.rules=rules
        ln=list(self.rules.keys())
        count=0
        states=[]
        for x in self.rules:
            count+=1
            print("Rule {} is: {}".format(count,x))
        moreStates=True
        while(moreStates):
            requestState=input("Would you like to add a halt state? y/n ")
            while(requestState.lower() not in ['y','n']):
                requestState=input("Would you like to an halt state? y/n ")
            if requestState.lower() =='n':
                moreStates=False
            elif requestState.lower()=='y':
                halt=int(input("What rule is your halt state? [1-{}]".format(count)))
                while (halt not in range(1,count)):
                    halt=int(input("What rule is your halt state? [1-{}]".format(count)))
                if halt not in states:
                    states.append(ln[halt-1])
                else:
                    pass       
        return states
    
    #Give the movement letter
    def letter(self,move):
        if movement ==-1:
            return 'L'
        elif movement ==1:
            return 'R'
        elif movement ==0:
            return 'Stay'
    
```

```python
#Values
accept=False
reject=False
turingSize=100
blankSpaces=2
FirstState='Start'
blankSymbol=' '

#Initialize a Turing machine
Turing=TM()
print("Initializing a new Turing Machine")
#Enter a Initial value to the Turing Machine
Init=str(input("Enter the initial value of the Turing machine:"))
Machine=Turing.create(Init,turingSize)

#Input symbols, blank space already included
symbols=Turing.symbols(Init)
#Input States, Start state already included
states=Turing.states()
#Input the rules for the Turing Machine, no rules included
rules=Turing.rules(symbols,states)
#Starting state of the Turing Machine
currentState=FirstState
#Starting symbol of the Turing Machine
currentSymbol=blankSymbol
#Starting location of the Turing Machine
head=0
#Accepted states of the Turing Machine
acceptStates=Turing.acceptStates(rules)

#Rejected states of the Turing Machine
rejectStates=Turing.rejectStates(rules)

#Halt state of the Turing Machine
haltState=Turing.haltState(rules)

print("\n"*10)
#Starting the Turing Machine
print("Starting Turing Machine: ")
Turing.printValues(Machine,head,blankSpaces)


Run=True
while Run:
    #Get the next state
    newState=Turing.newState(currentState,currentSymbol,rules)
    #Get the next Symbol
    newSymbol=Turing.newSymbol(currentState,currentSymbol,rules)
    #Get the next move
    movement=Turing.move(currentState,currentSymbol,rules)
    
    let=Turing.letter(movement)
    #Print the rule to do
    print("{}, '{}' -> {} ,'{}' ,{}".format(currentState,currentSymbol,newState,newSymbol,let))
    #Check the state at this spot
    checkState="{} {}".format(currentState,currentSymbol)
    
    #Apply the symbol to the location
    Machine[head]=newSymbol
    #Move the head to the next spot
    head+=movement
    #Configure the new state to current state
    currentState=newState
    #Read the symbol at the head
    currentSymbol=Machine[head]
    #Print the new values
    Turing.printValues(Machine,head,blankSpaces)
    
    #Apply if we halt
    if checkState in haltState:
        Run=False
        print("The Turing Machine Halted")
    else:
        Run=True
        
    if checkState in acceptStates and reject==False:
        accept=True
    elif checkState in rejectStates and accept==False:
        reject=True

if accept:
    print("The Turing Machine accepted")
elif reject:
    print("The Turing Machine rejected")    
```
